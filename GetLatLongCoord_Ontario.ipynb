```python
#Get Latitude, Longitude Coordinates with place names in CSV file
#CSV file should have full address with province, countryname to ensure data accuracy

import pandas as pd
from geopy import geocoders 
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut

#to ensure no timeout, change timeout period
geolocator = Nominatim(timeout=10)

#Find Lat,Long of Milton as Example
location = geolocator.geocode("Milton ONTARIO CANADA")

print('The coordinates of Milton,ON region are '+ str((location.latitude, location.longitude)))
```
